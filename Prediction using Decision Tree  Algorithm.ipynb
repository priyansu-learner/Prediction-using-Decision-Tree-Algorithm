### IMPORTING REQUIRED LIBRARIES

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### LOADING THE DATA

```python
data = pd.read_csv("iris.csv")
print(data.shape)
```

##### THERE ARE 150 ROWS AND 6 COLUMNS.

### CHECKING THE FIRST 5 ROWS OF THE DATA

```python
data.head()
```

### CHECKING FOR ANY NULL VALUE

```python
data.info()
```

##### NO NULL VALUE IS PRESENT AND ALL THE DATATYPES ARE CORRECT.

### DESCRIPTION OF THE DATA

```python
data.describe()
```

### SELECTING FEATURE AND TARGET VALUE

```python
x=data.iloc[:,[1,2,3,4]].values
y=data.iloc[:,5].values
print(x.shape)
print(y.shape)
```

### SPLITTING THE DATA INTO TRAINING AND TEST SET

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=1)
```

### CREATING CLASSIFIER

```python
from sklearn.tree import DecisionTreeClassifier
d_tree = DecisionTreeClassifier()
```

```python
d_tree.fit(x_train,y_train)
```

### TESTING THE CLASSIFIER

```python
y_pred=d_tree.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_pred,y_test)
print("Accuracy is ",accuracy)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_pred,y_test)
print("Confusion Matrix :\n",cm)
```

### VISUALIZING THE DECISION TREE

```python
from sklearn import tree
fig = plt.figure(figsize=(20,16))
_ = tree.plot_tree(d_tree,feature_names=x,class_names=y,filled=True)
```

```python
fig.savefig("decision_tree_iris.png")
```
